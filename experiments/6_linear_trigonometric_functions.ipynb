```python
import matplotlib.pyplot as plt
import torch

from blocks import SinBlock, CosBlock
from model import ModelFactory
```

```python
x = torch.arange(-10, 10, 0.1, dtype=torch.float32).unsqueeze(-1)
y = torch.zeros_like(x)
```

```python
model = ModelFactory(x, y, max_size=2).from_occurrence_dict({SinBlock: 1, CosBlock: 1})
print(model)
```

```python
model = ModelFactory(x, y, max_size=2).from_occurrence_dict({SinBlock: 1, CosBlock: 1})
plt.plot(x, y, color="orange", label="y=0")
plt.plot(x, model(x).detach().numpy(), color="blue", label="y = sin(x) + cos(x)")
plt.ylim((-1, 1))
plt.legend()
plt.show()
```
