```python
import matplotlib.pyplot as plt
import torch

from blocks import Log2Block, Log10Block, LnBlock
from model import ModelFactory
```

```python
x = torch.arange(-2, 6, 0.1, dtype=torch.float32).unsqueeze(-1)
y = torch.sigmoid(x)
```

```python
model = ModelFactory(x, y, max_size=2, log_mode="symmetric").from_class_list([Log2Block, Log10Block, LnBlock])
plt.scatter(x.numpy(), model(x).detach().numpy(), s=0.45)
plt.show()
```

```python
model = ModelFactory(x, y, max_size=2, log_mode="constant").from_class_list([Log2Block, Log10Block, LnBlock])
plt.scatter(x.numpy(), model(x).detach().numpy(), s=0.45)
plt.show()
```

```python
model = ModelFactory(x, y, max_size=2, log_mode="domain_adaptation").from_class_list([Log2Block, Log10Block, LnBlock])
plt.scatter(x.numpy(), model(x).detach().numpy(), s=0.45)
plt.show()
```
