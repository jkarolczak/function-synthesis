```python
import matplotlib.pyplot as plt
import torch

from blocks import CosBlock, SinBlock, Log2Block, LnBlock, Log10Block, LinearBlock
from model import ModelFactory, MultiLayerModelFactory
```

```python
torch.autograd.set_detect_anomaly(True)
x = torch.arange(1, 10, 0.1, dtype=torch.float32).unsqueeze(-1)
y = torch.sin(x) + torch.log2(x)
model = ModelFactory(x, y, max_size=2).from_class_list([CosBlock, SinBlock, Log2Block, LnBlock, Log10Block])
pred = model.forward(x)
print(torch.hstack([pred, y]))
print(model)
```

```python
plt.scatter(x, y)
plt.scatter(x, pred.detach().numpy(), marker='x', c='red')
plt.show()
```

```python
torch.autograd.set_detect_anomaly(True)
x = torch.arange(1, 50, 0.25, dtype=torch.float32).unsqueeze(-1)
y = torch.sin(x) + torch.log2(x) + torch.cos(x)
model = ModelFactory(x, y, max_size=5).from_class_list([CosBlock, SinBlock, Log2Block, LnBlock, Log10Block])
pred = model.forward(x)
print(torch.hstack([pred, y]))
for b, w in zip(model.blocks, model.scalar_weights):
    print(b, w)
```

```python
plt.scatter(x, y)
plt.scatter(x, pred.detach().numpy(), marker='x', c='red')
plt.show()
```
